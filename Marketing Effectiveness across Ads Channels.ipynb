# How does your advertising spend impact product sales? 
ðŸ“Š This visualization uncovers the relationship between proportional spending on platforms like TV, Social Media, and Affiliate Marketing and product sales. See which strategies drive results and where optimization is needed for maximum ROI! ðŸš€ </br>
- **Affiliate Marketing** shows a clear positive trend, indicating higher proportional spending correlates with increased product sales.
- **Billboards display** a slight positive relationship, suggesting moderate effectiveness.
- **Social Media and TV show** minimal correlation, indicating proportional spending has limited direct impact on product sales.
- **Google Ads** shows a slight negative trend, implying diminishing returns with higher proportional spending.
- **Influencer Marketing** has minimal impact, with a weak or inconsistent relationship to product sales.

```python
# Calculate the proportion of spend by platform
proportions = ads.iloc[:, :-1].div(ads.iloc[:, :-1].sum(axis=1), axis=0)
proportions['Product_Sold'] = ads['Product_Sold']

# Custom colors for each platform
colors = ['orange', 'red', 'pink', 'magenta', 'cyan', 'blue']

# Set up the plot grid
plt.figure(figsize=(15, 5))
for i, (column, color) in enumerate(zip(proportions.columns[:-1], colors)):
    plt.subplot(1, len(proportions.columns[:-1]), i + 1)
    sns.regplot(x=proportions[column], y=proportions['Product_Sold'], scatter_kws={'alpha': 0.6, 'color': color}, 
                line_kws={'color': color}, lowess=True)
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Product_Sold' if i == 0 else '')
    
    # Calculate and annotate correlation value
    corr_value = proportions[column].corr(proportions['Product_Sold'])
    plt.text(0.05, 0.95, f"Corr: {corr_value:.2f}", transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', color=color)

plt.suptitle('Proportional Spending vs. Product Sales with Correlations')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# Uncover the secrets behind advertising spend! 
ðŸš€ This chart highlights how platforms like TV, Google Ads, and Affiliate Marketing drive product sales. See the trends, identify diminishing returns, and optimize your strategy for maximum ROI! ðŸ’¡ </br>
- **Affiliate Marketing** demonstrates the strongest positive trend, suggesting it is highly effective at converting spend into product sales.
- **Billboards and Social Media** show moderate positive correlations, indicating consistent but less impactful performance.
- **Google Ads** exhibits a flatter trend, implying potential diminishing returns as spending increases.
- **TV shows** a weak positive trend, suggesting it may have limited direct impact on driving sales.
- **Influencer Marketing** has the least impactful trend, indicating minimal influence on product sales compared to other platforms.

```python
# Filter platforms based on correlation threshold
correlation_threshold = 0.1
correlations = ads.corr()['Product_Sold'].drop('Product_Sold')
high_correlation_platforms = correlations[abs(correlations) > correlation_threshold].index

# Custom colors for each platform (ensure the number of colors matches the platforms to be plotted)
colors = ['orange', 'red', 'pink', 'magenta', 'cyan', 'blue'][:len(high_correlation_platforms)]

# Plot Diminishing Returns Analysis with custom colors
plt.figure(figsize=(10, 6))
for platform, color in zip(high_correlation_platforms, colors):  # Use zip to iterate over platforms and colors together
    sns.regplot(
        x=ads[platform],
        y=ads['Product_Sold'],  # Use ads['Product_Sold'] directly for the dependent variable
        label=platform,
        scatter_kws={'alpha': 0.6, 'color': color},
        line_kws={'color': color},
        lowess=True
    )

# Add titles and labels
plt.title('Diminishing Returns Analysis (Filtered Platforms)')
plt.xlabel('Advertising Spend')
plt.ylabel('Products Sold')
plt.legend(title='Advertising Platforms')
plt.tight_layout()
plt.show()
```
